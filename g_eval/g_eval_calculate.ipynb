### 그냥 평균

```python
coherence = []
engagingness = []
naturalness = []
groundedness = []
```

```python
fp = open('g_eval_score_v2_persona.txt', 'r')
lines = fp.readlines()
for itr in lines:
   print(itr)
   if itr.startswith("## coherence score :"):
      coherence.append(itr)
   if itr.startswith("## engagingness score :"):
      engagingness.append(itr)
   if itr.startswith("## naturalness score :"):
      naturalness.append(itr)
   if itr.startswith("## groundedness score :"):
      groundedness.append(itr)
```

```python
coherence = [x.split(":  ")[1:] for x in coherence]
coherence = [eval(x[0][:-1]) for x in coherence]

engagingness = [x.split(":  ")[1:] for x in engagingness]
engagingness = [eval(x[0][:-1]) for x in engagingness]

naturalness = [x.split(":  ")[1:] for x in naturalness]
naturalness = [eval(x[0][:-1]) for x in naturalness]

groundedness = [x.split(":  ")[1:] for x in groundedness]
groundedness = [eval(x[0][:-1]) for x in groundedness]
```

```python
coherence_mean = [sum(x) / len(x) for x in coherence]
engagingness_mean = [sum(x) / len(x) for x in engagingness]
naturalness_mean = [sum(x) / len(x) for x in naturalness]
groundedness_mean = [sum(x) / len(x) for x in groundedness]
```

```python
print('Coherence arithmetic mean: ', sum(coherence_mean) / len(coherence_mean))
print('Engagingness arithmetic mean: ', sum(engagingness_mean) / len(engagingness_mean))
print('Naturalness arithmetic mean: ', sum(naturalness_mean) / len(naturalness_mean))
print('Groundedness arithmetic mean: ', sum(groundedness_mean) / len(groundedness_mean))
```
